```python
import gridfs

db = client['mydb']
# Set up GridFS
fs = gridfs.GridFS(db, collection='images')
```

```python
import os

# list folders
def list_folders_in_directory(directory):
    try:
        # Get the list of all files and directories in the specified directory
        items = os.listdir(directory)
        # Filter out only directories
        folders = [item for item in items if os.path.isdir(os.path.join(directory, item))]
        return folders
    except FileNotFoundError:
        return "Directory not found."
    except Exception as e:
        return str(e)

directory_path = 'F:/dataset/'
folders = list_folders_in_directory(directory_path)
print(folders)
```

```python
for folder_name in folders:
    # Path to local image directory
    image_directory = 'F:/dataset/{}/'.format(folder_name)
    
    # Counter to keep track of the number of images uploaded
    uploaded_count = 0
    max_images_per_folder = 200
    
    # Iterate over files in the directory
    for filename in os.listdir(image_directory):
        if filename.endswith(".jpg") or filename.endswith(".png"):
            with open(os.path.join(image_directory, filename), 'rb') as f:
                try:
                    # Store the file in GridFS with folder metadata
                    fs.put(f, filename=filename, metadata={'folder': folder_name, 'label': filename})
                    uploaded_count += 1
                    print(f"Uploaded {filename} to folder {folder_name}")
    
                    # Stop uploading after reaching the limit
                    if uploaded_count >= max_images_per_folder:
                        print(f"Reached the limit of {max_images_per_folder} images for folder {folder_name}")
                        break
                except Exception as e:
                    print(f"Error uploading {filename}: {e}")
```
